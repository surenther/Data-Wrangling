### Activity 3.01: Generating Statistics from a CSV File



```python
# Reading CSV

import pandas as pd

df = pd.read_csv("Boston_housing.csv") #Read csv
```

```python
# Displaying first 10 records 

print ("First 10 records")
df.head (10) #Display first 10 records
```

```python
# Displaying Total no of records
print ("Total Record count: " + str(len(df))) 
```

```python
# Removing columns from the dataframe 
columns_to_drop = ['CHAS','NOX','B','LSTAT'] #List of columns to drop
trimmed_df = df.drop(columns=columns_to_drop) #Drop using previous list

print ("Last 7 records")    
trimmed_df.tail(7) #Print last 7 records
```

```python
# Plotting histograms of all the columns in the DataFrame
import matplotlib.pyplot as plt

trimmed_df.hist(bins=10, figsize=(10, 10)) #Creating Histogram for all columns
plt.tight_layout() 
plt.show()
```

```python
# Plotting histograms of all the columns using for loop

for col in trimmed_df.columns:
    plt.figure()
    trimmed_df[col].hist() #Histogram creation
    plt.title(f"Histogram of {col}") #Set Title
    plt.xlabel(col) #Set X label
    plt.ylabel("Frequency") #Set Y label
    plt.show()
```

```python
# scatter plot of crime rate versus price
plt.scatter(trimmed_df['CRIM'], trimmed_df['PRICE'])

# Add labels and title
plt.xlabel("Crime Rate")
plt.ylabel("Price")
plt.title("Scatter Plot of Crime Rate vs. Price")

# Display the plot
plt.show()
```

```python
# Plot log10(crime) versus price
import numpy as np

# Apply log10 transformation to 'CRIM' column
trimmed_df['log_crime'] = np.log10(trimmed_df['CRIM'] + 1)  # Adding 1 to avoid log(0)

# scatter plot of crime rate versus price
plt.scatter(trimmed_df['log_crime'], trimmed_df['PRICE'])

# Add labels and title
plt.xlabel("log10 Crime Rate")
plt.ylabel("Price")
plt.title("Scatter Plot of Crime Rate vs. Price")

# Display the plot
plt.show()

```
