### Activity 3.01: Generating Statistics from a CSV File



```python
# Reading CSV

import pandas as pd

df = pd.read_csv("Boston_housing.csv") #Read csv
```

```python
# Displaying first 10 records 

print ("First 10 records")
df.head (10) #Display first 10 records
```

```python
# Displaying Total no of records
print ("Total Record count: " + str(len(df))) 
```

```python
# Removing columns from the dataframe 
columns_to_drop = ['CHAS','NOX','B','LSTAT'] #List of columns to drop
trimmed_df = df.drop(columns=columns_to_drop) #Drop using previous list

print ("Last 7 records")    
trimmed_df.tail(7) #Print last 7 records
```

```python
# Plotting histograms of all the columns in the DataFrame
import matplotlib.pyplot as plt

trimmed_df.hist(bins=10, figsize=(10, 10)) #Creating Histogram for all columns
plt.tight_layout() 
plt.show()
```

```python
# Plotting histograms of all the columns using for loop

for col in trimmed_df.columns:
    plt.figure()
    trimmed_df[col].hist() #Histogram creation
    plt.title(f"Histogram of {col}") #Set Title
    plt.xlabel(col) #Set X label
    plt.ylabel("Frequency") #Set Y label
    plt.show()
```

```python
# scatter plot of crime rate versus price
plt.scatter(trimmed_df['CRIM'], trimmed_df['PRICE'])

# Add labels and title
plt.xlabel("Crime Rate")
plt.ylabel("Price")
plt.title("Scatter Plot of Crime Rate vs. Price")

# Display the plot
plt.show()
```

```python
# Plot log10(crime) versus price
import numpy as np

# Apply log10 transformation to 'CRIM' column
trimmed_df['log_crime'] = np.log10(trimmed_df['CRIM'] + 1)  # Adding 1 to avoid log(0)

# scatter plot of crime rate versus price
plt.scatter(trimmed_df['log_crime'], trimmed_df['PRICE'])

# Add labels and title
plt.xlabel("log10 Crime Rate")
plt.ylabel("Price")
plt.title("Scatter Plot of Crime Rate vs. Price")

# Display the plot
plt.show()

```

```python
# Calculating statistics
mean_rooms = trimmed_df["RM"].mean()  # Mean rooms per dwelling
median_age = trimmed_df["AGE"].median()  # Median age of houses
mean_distance = trimmed_df["DIS"].mean()  # Mean distances to five employment centers

# Percentage of houses with price < $20,000
low_price_percentage = (trimmed_df["PRICE"] < 20).mean() * 100  # Convert to percentage

# Print results
print(f"Mean Rooms per Dwelling: {mean_rooms:.2f}")
print(f"Median Age of Houses: {median_age:.2f}")
print(f"Mean Distance to Employment Centers: {mean_distance:.2f}")
print(f"Percentage of Houses Priced Below $20,000: {low_price_percentage:.2f}%")
```

### Activity 4.01: Working with the Adult Income Dataset 



```python
# Reading data in CSV
df = pd.read_csv("adult_income_data.csv") #Read csv
df.head (10) # Display first 10 records
```

```python
#Reading column title from the text file

#Creating function for reading text file
def read_text_file(file_path):
    name = []
    try:
        with open (file_path,'r') as file:
            for line in file:
                column_name =line.split(':')[0].strip() #Spliting the value and removing linefeed/spaces
                if column_name: 
                    name.append(column_name) #appending to pythong list
            return name
    except Exception as e:
        print("An Error Ocurred: {e}")

title = read_text_file ("adult_income_names.txt")
print (title)
```

```python
#Append Income title
title.append('Income')
print (title)
```

```python
#Giving column title to the data frame
df.columns = title
df.head (10)
```

```python
#Finding missing value
missing_val = df.isna().sum()
print ("The Total Missing values \n")
print(missing_val)
```

```python
#Select Age, Education and occupation
selected_df = df[['age','education','occupation']]
selected_df.head (10)
```

```python
#plot historgram with bin size 20
plt.hist(selected_df['age'],bins=range(0,105,20),edgecolor='black')

# Add labels and title
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.title("Age Distribution (Bin Size = 20)")

# Show the plot
plt.show()
```

```python
# Function to strip whitespace
def strip_whitespace(value):
    if isinstance(value, str):  # Ensure it's a string before stripping
        return value.strip()
    return value
```

```python
# Apply the function to all string columns

for col in df.select_dtypes(include=['object']).columns:   # Select String column
    df[col + '_temp'] = df[col].apply(strip_whitespace)  #Apply whitespace function & create _temp column
    df[col] = df[col + '_temp'] # Replace original column with temp
    df.drop(columns=[col + '_temp'], inplace=True) # drop the temp columns
    
df.head(10)
```

```python
# The number of people who are aged between 30 and 50

num_people = df[(df['age'] >= 30) & (df['age'] <= 50)].shape[0] #[shape] returns the no of rows that match the condition
print(f"Number of people aged between 30 and 50: {num_people}")
```

```python
# Grouping based on age and education

mean_age_distribution = df.groupby("education")["age"].mean()
print(mean_age_distribution)
```

```python
# Group by occupation and get summary statistics for age
occupation_age_stats = df.groupby("occupation")["age"].describe()
print(occupation_age_stats)
print("\n")

# Find the occupation with the oldest average age
oldest_avg_occupation = occupation_age_stats["mean"].idxmax()  # idxmax will get us max mean value
print(f"Occupation with the oldest workers on average: {oldest_avg_occupation}")

# Calculate the 75th percentile of age for each occupation
age_75th_percentile = df["age"].quantile(0.75)

# Find the occupation with the highest proportion of workers above the 75th percentile
df_above_75th = df[df["age"] > age_75th_percentile]
largest_share_occupation = df_above_75th["occupation"].value_counts(normalize=True).idxmax()

print(f"Occupation with the largest share of workforce above 75th percentile: {largest_share_occupation}")
```

```python
#Outliers detection using Subset and groupby
occupation_stats=selected_df.groupby('occupation').describe()['age'] #finding the stats for occupation based on age

#Ploting outliers

# Set figure size
fig, ax = plt.subplots(figsize=(15, 8))

# Create horizontal bar chart
ax.barh(occupation_stats.index, occupation_stats['count'], color='lightblue')

# Labels and appearance
ax.set_yticks(range(len(occupation_stats.index)))
ax.set_yticklabels(occupation_stats.index, fontsize=13)
ax.set_xlabel("Count", fontsize=14)
ax.set_title("Count of Each Occupation", fontsize=16)

# Show plot
plt.show()
```

```python
# Merge the two Dataframe

# Create 2 dataframes
df_1 = df[['age','workclass','occupation']]
df_2 = df[['education','occupation','marital-status']]

#Merge & Print
df_merged = pd.merge(df_1,df_2,on='occupation',how='inner').drop_duplicates()
df_merged.head(10)    
```
